# Define functions

```python
def plotExpPeaks(gene,exp_df,peak_df,inhib_effect_cutoff,dirname=""):
    plt.clf()
    
    ax = sns.lineplot(data=exp_df,x='fragment center (aa)',y='E = inhibitory effect (enrichment)',alpha=.75)
    ax.axhline(inhib_effect_cutoff,c='r',ls='--')

    plt.hlines(peak_df['peak_fragment_score'],peak_df['peak region first fragment center (aa)'],peak_df['peak region last fragment center (aa)'],
              colors='g')

    fcenter = peak_df['fragment center (aa)']
    wcontacts = peak_df['peak_fragment_score']
    ax = sns.scatterplot(x=fcenter,y=wcontacts,color='r')
    ax.set_title(gene)
    
    filename = f"{gene}_experimentalpeaks.png" if dirname == "" else f"{dirname}/{gene}_experimentalpeaks.png"
    plt.savefig(filename,dpi=300)
    
    plt.show()
```

```python
gene_30_aa = {'EGFP-coding-pEGFP',
             'folA-coding-EcoliBL21DE3',
             'ftsZ-coding-EcoliBL21DE3',
             'groL-coding-EcoliBL21DE3',
             'groS-coding-EcoliBL21DE3',
             'gyrA-coding-EcoliBL21DE3',
             'ileS-coding-EcoliBL21DE3',
#              'lptG-coding-EcoliBL21DE3',
             'rpIL-coding-EcoliBL21DE3',
             'rpoB-coding-EcoliBL21DE3',
             'ssb-coding-EcoliBL21DE3'}

# gene2cutoff = {
#      'EGFP-coding-pEGFP':-3.0,
#      'folA-coding-EcoliBL21DE3':-3.0,
#      'ftsZ-coding-EcoliBL21DE3':-3.0,
#      'groL-coding-EcoliBL21DE3':-1.5,
#      'groS-coding-EcoliBL21DE3':-2.0,
#      'gyrA-coding-EcoliBL21DE3':-3.5,
#      'ileS-coding-EcoliBL21DE3':-3.5,
#      'lptG-coding-EcoliBL21DE3':-1.5,
#      'rpIL-coding-EcoliBL21DE3':-3.0,
#      'rpoB-coding-EcoliBL21DE3':-3.0,
#      'ssb-coding-EcoliBL21DE3':-3.5,
# }

def autoCutoff(filt_df):
    filt_df = filt_df[filt_df['E = inhibitory effect (enrichment)']>-2.0]
    print(f"mean: {filt_df['E = inhibitory effect (enrichment)'].mean()}")
    print(f"std: {filt_df['E = inhibitory effect (enrichment)'].std()}")
    std_weight = 2.5
    return filt_df['E = inhibitory effect (enrichment)'].mean() - std_weight*filt_df['E = inhibitory effect (enrichment)'].std()

# cutoff = -3.0
lookaround_distance = 30
grouping_distance = 25
max_gap_distance = 5

dirname = f"experimentalpeaks_variablecutoff_modeabsolute_grouping{grouping_distance}_maxgapdistance{max_gap_distance}"
os.makedirs(dirname,exist_ok=True)

peak_df_list = []
for gene,grouped_exp_df in exp_df.groupby('gene'):
    frag_len = 30 if gene in gene_30_aa else 14
    filt_exp_df = exp_df[(exp_df['gene']==gene)&(exp_df['fragment length (aa)']==frag_len)]
    cutoff = autoCutoff(filt_exp_df)
    print(gene,cutoff)
    peak_df,peak_score_list = peakFindingFromExpData(filt_exp_df,'E = inhibitory effect (enrichment)',
                                    exp_val_threshold=cutoff,
                                    lookaround_distance=lookaround_distance,
                                    grouping_distance=grouping_distance,
                                    max_gap_distance=max_gap_distance)

    plotExpPeaks(gene,filt_exp_df,peak_df,cutoff,dirname)
    
#     plt.clf()
#     sns.lineplot(x=filt_exp_df['fragment center (aa)'],y=peak_score_list,color='r')
#     plt.show()
    peak_df['inhib_effect_cutoff'] = cutoff
    peak_df_list.append(peak_df)

    
all_exp_peaks_df = pd.concat(peak_df_list,ignore_index=True)
print(f"{len(all_exp_peaks_df)} experimental fragment peaks total")
all_exp_peaks_df.to_csv(f"230223_experimentalpeaks_variablecutoff_modeabsolute_grouping{grouping_distance}_maxgapdistance{max_gap_distance}.csv")
```
